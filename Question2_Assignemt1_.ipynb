# Question 2.1 The first step is to find out the “the intersection of the curve and the line” like below graph

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Define the graph function
def graph(formula1, formula2, x_range):
    # Convert x_range to a numpy array
    x = np.array(x_range)

    # Use x as the range variable and compute y-values
    y1 = formula1(x)
    y2 = formula2(x)

    # Plot both functions
    plt.plot(x, y1, label="Formula 1")
    plt.plot(x, y2, label="Formula 2")

    # Fill the area between y1 and y2 where y1 < y2
    plt.fill_between(x, y1, y2, where=(y1 < y2), color='red', alpha=0.3)

    # Set the y-limits for the plot
    plt.ylim(ymin=0)

    # Add labels and legend
    plt.legend()

    # Show the plot
    plt.show()

    # Find the intersections using fsolve
    def difference(x):
        return formula1(x) - formula2(x)

    # Use fsolve to find the x-values where the two functions intersect
    x_intersections = fsolve(difference, [-2, 6])  # Adjust initial guesses [-2, 6] as needed

    # Calculate the corresponding y-values for the intersections
    y_intersections = formula1(x_intersections)

    # Print the intersection points as tuples (x, y)
    print("Intersection points:")
    for x_int, y_int in zip(x_intersections, y_intersections):
        print(f"({x_int:.3f}, {y_int:.3f})")
```

```python
formula1 = lambda x: 10 - 2*x
formula2 = lambda x: 5 + 4*x - x**2

# Call the graph function with the specified formulas and x range
graph(formula1, formula2, np.arange(-2, 7, 0.0001))
```
